```python
import os, cv2, numpy as np, pytesseract, matplotlib.pyplot as plt
from skimage.restoration import richardson_lucy
from skimage.metrics import peak_signal_noise_ratio
```

```python
ROOT="/content/drive/MyDrive/Datasets"

BLUR=f"{ROOT}/blurred"
SEMI=f"{ROOT}/semi-blurred"
CLEAN=f"{ROOT}/clean"

OUT=f"{ROOT}/results"
DEBLUR=f"{OUT}/deblurred"
PLATE=f"{OUT}/plates"
OCR=f"{OUT}/ocr"

for p in [DEBLUR,PLATE,OCR]:
    os.makedirs(p,exist_ok=True)
```

```python
plate_cascade = cv2.CascadeClassifier(cv2.data.haarcascades+"haarcascade_russian_plate_number.xml")
```

```python
def deblur(img):

    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

    psf=np.ones((5,5))/25

    restored=richardson_lucy(gray/255.,psf,10)

    restored=(restored*255).clip(0,255).astype(np.uint8)

    return cv2.cvtColor(restored,cv2.COLOR_GRAY2BGR)
```

```python
clean_imgs=[cv2.imread(f"{CLEAN}/{f}") for f in os.listdir(CLEAN)]
```

```python
inputs=os.listdir(BLUR)+os.listdir(SEMI)

for i,f in enumerate(inputs):

    path=f"{BLUR}/{f}" if f in os.listdir(BLUR) else f"{SEMI}/{f}"

    img=cv2.imread(path)

    restored=deblur(img)

    cv2.imwrite(f"{DEBLUR}/{f}",restored)

    ref=clean_imgs[np.random.randint(len(clean_imgs))]
    ref=cv2.resize(ref,(restored.shape[1],restored.shape[0]))

    psnr=peak_signal_noise_ratio(ref,restored)

    gray=cv2.cvtColor(restored,cv2.COLOR_BGR2GRAY)

    plates=plate_cascade.detectMultiScale(gray,1.1,4)

    text="No plate"

    for (x,y,w,h) in plates[:1]:
        plate=restored[y:y+h,x:x+w]
        cv2.imwrite(f"{PLATE}/{f}",plate)
        text=pytesseract.image_to_string(plate,config="--psm 7")
        open(f"{OCR}/{f}.txt","w").write(text)

    print(f"{i+1}. {f} PSNR {round(psnr,2)} OCR {text.strip()}")

    plt.imshow(cv2.cvtColor(restored,cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
```

```python
import torch
import torch.nn as nn

class UNet(nn.Module):
    def __init__(self):
        super().__init__()

        def C(in_c,out_c):
            return nn.Sequential(
                nn.Conv2d(in_c,out_c,3,1,1),
                nn.ReLU(),
                nn.Conv2d(out_c,out_c,3,1,1),
                nn.ReLU()
            )

        self.d1=C(3,64)
        self.d2=C(64,128)
        self.d3=C(128,256)
        self.pool=nn.MaxPool2d(2)

        self.mid=C(256,512)

        self.u3=C(512+256,256)
        self.u2=C(256+128,128)
        self.u1=C(128+64,64)

        self.up=nn.Upsample(scale_factor=2)
        self.out=nn.Conv2d(64,3,1)

    def forward(self,x):
        d1=self.d1(x)
        d2=self.d2(self.pool(d1))
        d3=self.d3(self.pool(d2))
        m=self.mid(self.pool(d3))

        u3=self.u3(torch.cat([self.up(m),d3],1))
        u2=self.u2(torch.cat([self.up(u3),d2],1))
        u1=self.u1(torch.cat([self.up(u2),d1],1))

        return self.out(u1)
```

```python
import glob
from torch.utils.data import Dataset, DataLoader

class DeblurDataset(Dataset):

    def __init__(self):

        self.clean = sorted(glob.glob(CLEAN+"/*.jpg")+glob.glob(CLEAN+"/*.png"))

        self.blur = sorted(glob.glob(BLUR+"/*.jpg")+glob.glob(BLUR+"/*.png"))
        self.semi = sorted(glob.glob(SEMI+"/*.jpg")+glob.glob(SEMI+"/*.png"))

        # merge blurred + semi-blurred
        self.inputs = self.blur + self.semi

        print("Clean:",len(self.clean))
        print("Blurred:",len(self.blur))
        print("Semi:",len(self.semi))
        print("Total Inputs:",len(self.inputs))

        if len(self.inputs)==0:
            raise ValueError("NO BLURRED / SEMI IMAGES FOUND")

        if len(self.clean)==0:
            raise ValueError("NO CLEAN IMAGES FOUND")

    def __len__(self):
        return len(self.inputs)

    def __getitem__(self, idx):

        x = cv2.imread(self.inputs[idx])
        y = cv2.imread(self.clean[idx % len(self.clean)])

        if x is None or y is None:
            raise ValueError("Image read failed")

        x=cv2.resize(x,(256,256))
        y=cv2.resize(y,(256,256))

        x=x.astype(np.float32)/255.0
        y=y.astype(np.float32)/255.0

        return torch.from_numpy(x).permute(2,0,1), \
               torch.from_numpy(y).permute(2,0,1)
```

```python
import torch.nn as nn

class UNet(nn.Module):

    def __init__(self):
        super().__init__()

        def C(in_c,out_c):
            return nn.Sequential(
                nn.Conv2d(in_c,out_c,3,1,1),
                nn.ReLU(),
                nn.Conv2d(out_c,out_c,3,1,1),
                nn.ReLU()
            )

        self.d1=C(3,64)
        self.p1=nn.MaxPool2d(2)

        self.d2=C(64,128)
        self.p2=nn.MaxPool2d(2)

        self.b=C(128,256)

        self.u2=nn.ConvTranspose2d(256,128,2,2)
        self.c2=C(256,128)

        self.u1=nn.ConvTranspose2d(128,64,2,2)
        self.c1=C(128,64)

        self.out=nn.Conv2d(64,3,1)

    def forward(self,x):

        d1=self.d1(x)
        d2=self.d2(self.p1(d1))

        b=self.b(self.p2(d2))

        u2=self.u2(b)
        u2=torch.cat([u2,d2],1)
        u2=self.c2(u2)

        u1=self.u1(u2)
        u1=torch.cat([u1,d1],1)
        u1=self.c1(u1)

        return torch.sigmoid(self.out(u1))
```

```python
device="cuda" if torch.cuda.is_available() else "cpu"

model=UNet().to(device)

criterion=nn.L1Loss()
optimizer=torch.optim.Adam(model.parameters(),lr=1e-4)

print("Device:",device)
```

```python
EPOCHS=15

for epoch in range(EPOCHS):

    model.train()
    total=0

    for x,y in loader:

        x=x.to(device)
        y=y.to(device)

        pred=model(x)

        loss=criterion(pred,y)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total+=loss.item()

    print(f"Epoch {epoch+1}/{EPOCHS} Loss:",round(total/len(loader),4))
```

```python
import os
import glob
import cv2
import numpy as np # Ensure numpy is imported for checking np.ndarray
from torch.utils.data import Dataset, DataLoader
import torch

class DeblurDataset(Dataset):

    def __init__(self):
        # Initial globbing to find all potential image files
        all_clean_candidates = sorted(glob.glob(CLEAN + "/*.jpg") + glob.glob(CLEAN + "/*.png"))
        all_blur_candidates = sorted(glob.glob(BLUR + "/*.jpg") + glob.glob(BLUR + "/*.png"))
        all_semi_candidates = sorted(glob.glob(SEMI + "/*.jpg") + glob.glob(SEMI + "/*.png"))

        # Filter out corrupted or invalid clean images
        self.clean_files = []
        for f_path in all_clean_candidates:
            try:
                img = cv2.imread(f_path)
                # A valid image will be a numpy array with at least 2 dimensions (height, width) and a positive size
                if isinstance(img, np.ndarray) and img.size > 0 and img.shape[0] > 0 and img.shape[1] > 0:
                    self.clean_files.append(f_path)
                else:
                    print(f"Warning: Skipping invalid or empty clean image: {f_path} (shape: {img.shape if isinstance(img, np.ndarray) else 'None'}, size: {img.size if isinstance(img, np.ndarray) else 'None'})")
            except Exception as e:
                print(f"Warning: Skipping clean image {f_path} due to error during read: {e}")

        # Filter out corrupted or invalid blurred images
        filtered_blur_files = []
        for f_path in all_blur_candidates:
            try:
                img = cv2.imread(f_path)
                if isinstance(img, np.ndarray) and img.size > 0 and img.shape[0] > 0 and img.shape[1] > 0:
                    filtered_blur_files.append(f_path)
                else:
                    print(f"Warning: Skipping invalid or empty blurred image: {f_path} (shape: {img.shape if isinstance(img, np.ndarray) else 'None'}, size: {img.size if isinstance(img, np.ndarray) else 'None'})")
            except Exception as e:
                print(f"Warning: Skipping blurred image {f_path} due to error during read: {e}")

        # Filter out corrupted or invalid semi-blurred images
        filtered_semi_files = []
        for f_path in all_semi_candidates:
            try:
                img = cv2.imread(f_path)
                if isinstance(img, np.ndarray) and img.size > 0 and img.shape[0] > 0 and img.shape[1] > 0:
                    filtered_semi_files.append(f_path)
                else:
                    print(f"Warning: Skipping invalid or empty semi-blurred image: {f_path} (shape: {img.shape if isinstance(img, np.ndarray) else 'None'}, size: {img.size if isinstance(img, np.ndarray) else 'None'})")
            except Exception as e:
                print(f"Warning: Skipping semi-blurred image {f_path} due to error during read: {e}")

        # Combine filtered blurred and semi-blurred for inputs
        self.all_input_files = filtered_blur_files + filtered_semi_files

        if not self.all_input_files:
            raise RuntimeError("No valid input images found in BLUR or SEMI directories after filtering.")
        if not self.clean_files:
            raise RuntimeError("No valid clean images found in CLEAN directory after filtering.")

    def __len__(self):
        return len(self.all_input_files)

    def __getitem__(self, idx):
        # Load input image (already validated during init)
        inp_path = self.all_input_files[idx]
        inp = cv2.imread(inp_path) # Should now always be valid

        # Load target image (already validated during init)
        tgt_path = self.clean_files[idx % len(self.clean_files)]
        tgt = cv2.imread(tgt_path) # Should now always be valid

        # Minimal checks in case of very rare runtime issues or concurrent file deletion
        if inp is None or tgt is None:
             raise RuntimeError(f"Error loading image during __getitem__ for {inp_path} or {tgt_path}, despite pre-filtering. This should not happen.")

        inp = cv2.resize(inp,(256,256))
        tgt = cv2.resize(tgt,(256,256))

        inp = torch.tensor(inp).permute(2,0,1).float()/255.
        tgt = torch.tensor(tgt).permute(2,0,1).float()/255.

        return inp, tgt
```
